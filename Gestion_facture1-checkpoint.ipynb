```python
from tkinter import *
import tkinter.ttk as ttk

import mysql.connector
#class facture
class facture:
    def __init__(self,id=0,id_client=0,num_telephone=0,num_ticket=0,date='0/0/0'):
        self.id=id
        self.id_client=id_client
        self.num_telephone=num_telephone
        self.num_ticket=num_ticket
        self.date=date
        
        self.db=mysql.connector.connect(
        host='localhost',
        user='rym',
        password='12345',
        database='gestion_patisserie'
        )
        self.mycursor=self.db.cursor()
        
    def ajouterfact(self):
        sql = 'INSERT INTO facture (id, id_client, num_telephone, num_ticket, date) VALUES (%s, %s, %s, %s, %s)'
        val=(self.id,self.id_client,self.num_telephone,self.num_ticket,self.date)
        self.mycursor.execute(sql,val)
        self.db.commit()
        print(self.mycursor.rowcount,'record inserted.')   
    def afficherfact(self):
        self.mycursor.execute("SELECT * FROM facture")
        resultat=self.mycursor.fetchall()
        return resultat      
    def supprimerfact(self,num):
        sql = "DELETE FROM facture WHERE id = %s"
        val = (num,)
        
        self.mycursor.execute(sql, val)
        self.db.commit()
        print(self.mycursor.rowcount,'record deleted.')
    def modifierfacture(self, id_0, id_client, num_telephone, num_ticket, date):
        sql = "Update facture set id_client=%s,num_telephone=%s,num_ticket=%s,date=%s where id=%s"
        val = (id_client,num_telephone,num_ticket,date, id_0)
        self.mycursor.execute(sql, val)
        self.db.commit()
        print(self.mycursor.rowcount, "record updated.")#optionnell
        
    
class Gestionfact():
    def __init__(self,root):
        self.root = root
        self.root.geometry('600x600')
        self.root.configure(background='light green')
        self.root.title("Gestion fact")
        
        
        self.id=IntVar()
        self.id_client=IntVar()
        self.num_telephone=IntVar()
        self.num_ticket=IntVar()
        self.date=StringVar()
        
       
        #Formulaire ajout Facture dans __init():
        #============================================================================
        
        id_emprunt_lbl = Label(self.root,text = "id",anchor='w', bg="light green")
        id_emprunt_lbl.grid(row = 1,column = 0,padx = 40,pady = 40)
        id_emprunt_field = Entry(self.root,textvariable = self.id)
        id_emprunt_field.grid(row = 1,column = 1,ipady = 7,ipadx = 20,padx = 20)
        #=============================================================================
        
        date_debut_lbl = Label(self.root,text = "id de client",anchor='w', bg="light green")
        date_debut_lbl.grid(row = 2,column = 0,padx = 40,pady = 40)
        date_debut_field = Entry(self.root,textvariable = self.id_client)
        date_debut_field.grid(row = 2,column = 1,ipady = 7,ipadx = 20,padx = 20)
        
         #============================================================================
        
        id_emprunt_lbl = Label(self.root,text = "numero de telephone",anchor='w', bg="light green")
        id_emprunt_lbl.grid(row = 3,column = 0,padx = 40,pady = 40)
        id_emprunt_field = Entry(self.root,textvariable = self.num_telephone)
        id_emprunt_field.grid(row = 3,column = 1,ipady = 7,ipadx = 20,padx = 20)
        
         #============================================================================
        
        id_emprunt_lbl = Label(self.root,text = "numero de ticket",anchor='w', bg="light green")
        id_emprunt_lbl.grid(row = 4,column = 0,padx = 40,pady = 40)
        id_emprunt_field = Entry(self.root,textvariable = self.num_ticket)
        id_emprunt_field.grid(row = 4,column = 1,ipady = 7,ipadx = 20,padx = 20)
        
        #=============================================================================
        
        type_lbl = Label(self.root,text="Date de facture",anchor='w', bg="light green")
        type_lbl.grid(row = 5,column = 0,pady = 40)
        type_field = Entry(self.root,textvariable = self.date)
        type_field.grid(row = 5,column = 1,ipady = 7,ipadx = 20,padx = 20)
    
        #=====================Boutton ajout
        ajouter_emprunt_btn = Button(self.root,text = "Ajouter",command = self.add,anchor='c',bg="pink")
        ajouter_emprunt_btn.grid(row = 1,column = 6,ipady = 4,ipadx = 60,pady = 5)
          
        #=====================Boutton affichage à ajouter après boutton "Ajouter" dans __init()__
        afficheremprunt_btn = Button(self.root,text = "Afficher",command = self.view1,anchor='c',bg="pink")
        afficheremprunt_btn.grid(row = 2,column = 6,ipady = 4,ipadx = 60,pady = 5)
        
        #=====================Boutton Supprimer à ajouter après boutton "Afficher" dans __init()__
        supp_etudiant_btn = Button(self.root,text = "Supprimer",command = self.remove,anchor='c',bg="pink")
        supp_etudiant_btn.grid(row = 3,column = 6,ipady = 4,ipadx = 60,pady = 5)
        
        #=====================Boutton affichage à modifier après boutton "Ajouter" dans __init()__
        affich_facture_btn = Button(self.root,text = "Modifier",command = self.update,anchor='c',bg="pink")
        affich_facture_btn.grid(row = 4,column = 6,ipady = 4,ipadx = 60,pady = 5)
        
    #Fonction d'ajout (sera appelée dérière le boutton "Ajouter"
    def add(self):
        E1 = facture(int(self.id.get()),self.id_client.get(),self.num_telephone.get(),self.num_ticket.get(),self.date.get())
        print("facture: ",E1.id)
        E1.ajouterfact()   
    def view1(self):
        #==========================Show Frame
        self.root=Tk()
        self.root.geometry('700x300')
        self.root.title("Gestion des factures")
        show_frame = Frame(self.root)
        show_frame.place(width = 800,x = 0,y = 0 ,height = 100)
        labl_show = Label(show_frame,text = "Liste des factures")
        labl_show.pack()
        
        #========================Main Frame
        
        main_frame = Frame(self.root,bd = 10,relief = SUNKEN)
        main_frame.place(width = 600,height = 200,x = 8,y = 58)
        tree = ttk.Treeview(main_frame,height = 200)
        vsb = ttk.Scrollbar(main_frame,command = tree.yview,orient = "vertical")
        tree.configure(yscroll = vsb.set)
        vsb.pack(side = RIGHT,fill = Y)
        tree.pack(side = TOP,fill = X)
        
        tree['columns'] = ("1","2","3","4","5","6")
        
        tree.column('#0',width=50)
        tree.column('1',width=80)
        tree.column('2',width=80)
        tree.column('3',width=80)
        tree.column('4',width=80)
        
        
        tree.heading("#0",text = "",anchor='w')
        tree.heading("1",text = "id",anchor='w')
        tree.heading("2",text = "id_client",anchor='w')
        tree.heading("3",text = "num_telephone",anchor='w')
        tree.heading("4",text = "num_ticket",anchor='w')
        tree.heading("5",text = "date",anchor='w')
       
        E1=facture()
        rows=E1.afficherfact()
        j=1
        for i in rows:
            tree.insert("","end",text=str(j), values = (f'{i[0]}',f'{i[1]}',f'{i[2]}',f'{i[3]}',f'{i[4]}'))
            j+=1
            
    #Fonction dse suppression d'un étudiant sera appelée dans le boutton "Supprimer"  
    def remove(self):
        E1 =facture('%s')
        E1.supprimerfact(self.id.get())
    #Fonction dse suppression d'un étudiant sera appelée dans le boutton "Supprimer"  
    def update(self):
        F = facture(self.id.get(),self.id_client.get(),self.num_telephone.get(),self.num_ticket.get(),self.date.get())
        #print("Stock: ",s.id)
        F.modifierfacture(self.id.get(), self.id_client.get(), self.num_telephone.get(), self.num_ticket.get(),self.date.get())

#if __name__ == "__main__":
root = Tk()
l = Gestionfact(root)
root.mainloop()
```
